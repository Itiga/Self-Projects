Question 7 . Email Spam Detection Model Using Logistic Regression

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
daf = pd.read_csv("/content/Email_Dataset.csv")
print(daf)
```

```python
df = daf.where((pd.notnull(daf)), '')
print(df)
```

```python
df.loc[df['CATEGORY'] == 'Spam' , 'CATEGORY',] = 0
df.loc[df['CATEGORY'] == 'Not Spam' , 'CATEGORY',] = 1
```

```python
X = df['MESSAGE']
Y = df['CATEGORY']
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=43)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')

print(X_train)
print(X_train_features)
```

```python
model = LogisticRegression()
model.fit(X_train_features,Y_train)
```

```python
pred_train = model.predict(X_train_features)
acc_train = accuracy_score(Y_train,pred_train)
print("Accuracy on training data:",acc_train)
```

```python
pred_test = model.predict(X_test_features)
acc_test = accuracy_score(Y_test,pred_test)
print("Accuracy on test data:",acc_test)
```

Accuracy on training data - Accuracy on test data < 1. This implies there are no problems of overfitting.

## Prediction using the model
